```python
# Assignment 6a
# Crystal Worley-Criado

import arcpy

# Select by Attribute

arcpy.env.workspace = r"C:\Users\cryst\gis_programming\assignment_1b\Admin\AdminData.gdb"

flayer = arcpy.MakeFeatureLayer_management("U.S. Cities","Cities_Layer")

qry = "POP1990 > 20000"
arcpy.management.SelectLayerByAttribute(flayer,"NEW_SELECTION",qry)

my_count = arcpy.management.GetCount(flayer)
print(f"Number of Currently Selected Cities: {my_count}")
```

```python
# Select by Location

arcpy.management.SelectLayerByLocation(flayer,"WITHIN_A_DISTANCE","U.S. Rivers (Generalized)","10 miles")

my_count = arcpy.management.GetCount(flayer)
print(f"Number of Currently Selected Cities: {my_count}")
```

```python
# Select cities that have a population greater than 30000 and are within 10 miles of a river

# 30000

qry = "POP1990 > 30000"
arcpy.management.SelectLayerByAttribute(flayer,"NEW_SELECTION",qry)

my_count = arcpy.management.GetCount(flayer)
print(f"Number of Currently Selected Cities: {my_count}")
```

```python
# within 10 miles of a river, make a subset selection of the current selections

arcpy.management.SelectLayerByLocation(flayer,"WITHIN_A_DISTANCE","U.S. Rivers (Generalized)","10 miles","SUBSET_SELECTION")

my_count = arcpy.management.GetCount(flayer)
print(f"Number of Currently Selected Cities: {my_count}")
```

```python
# Count the total population of cities with more than 30000 people within 10 miles of a river

field = 'POP1990'
total = 0
i = 1

with arcpy.da.SearchCursor(flayer,field) as cursor:
    for row in cursor:
        print(i, str(row[0]))
        total = total + row[0]
        i = i + 1
        
print(f"Total population is: {total:,}")
```
